```python
from graphviz import Graph, Digraph
```

```python
g = Graph()
g.node('A', 'Madison', color='red')
g.node('B', 'Chicago')
g.edge('A', 'B', label = 'Tolls+Short', color='red')
g.edge('A', 'B', label = 'Cheap+Slow', penwidth='3')
g.edge('A', 'A')
g
```

```python
g = Digraph()
#only a "cycle" if it loops in one direction back to itself
g.edge('A', 'B', penwidth='3', color='red')
g.edge('B', 'C')
g.edge('C', 'D')
g.edge('D', 'A', penwidth='3', color='red')
g.edge('X', "Y") #these are not "connected"
g.edge("Y", "X")
g
```
