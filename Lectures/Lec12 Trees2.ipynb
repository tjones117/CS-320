# root: node without parents
# leaf: node without children
# tree: has exactly one root and every node has one parent
# binary tree: tree with every node having no more than two children

```python
#binary tree
class Node:
    def __init__(self, val):
        self.val = val

        #general tree
        #self.children = []
        
        #binary tree
        self.left = None
        self.right = None
        
    def dump(self):
        print(self.val)
        if self.left != None:
            self.left.dump()
        if self.right != None:
            self.right.dump()
            
    def _repr_svg_(self, g=None):
        if g == None:
            gv = Digraph()
        else:
            gv = g
            
        #draw myself
        gv.node(self.val)
        #draw children (recursive)
        if self.left != None:
            self.left._repr_svg_(gv)
            gv.edge(self.val, self.left.val, label="L")
        if self.right != None:
            self.right._repr_svg_(gv)
            gv.edge(self.val, self.right.val, label="R")
            
        return gv._repr_svg_()
    
    def contains(self, val):
        ##Print subtree we're working on
        display.myself ###FIX THIS
        
        #YES if any of the following
        #I have value
        #left or right subtree has value
        if self.val == val:
            return True
        
        if self.left != None and self.left.contains(val):
            return True
        
        if self.right != None and self.right.contains(val):
            return True
        
        return False
        
        
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.right = Node("D")
root.left.left = Node("E")
root.contains("E")
```
